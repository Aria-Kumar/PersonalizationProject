```python
import os 
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

```python
os.chdir('D:\\Columbia University\\Personalization\\Project\\lastfm-dataset-1K')
```

```python
factors = pd.read_table('factor.csv', header=0, sep = ',')
learning = pd.read_table('learning.csv', header=0, sep = ',')
reg = pd.read_table('reg.csv', header=0, sep = ',')
```

```python
factors_regplot = sns.regplot(x='Factors',y='RMSE', data=factors)
factors_regplot.axes.set_title('Change in Factors')
```

```python
factors_plot = sns.barplot(x="Factors", y="RMSE", data=factors, color = 'blue', saturation=.3)
factors_plot.axes.set_title('RMSE vs. Change in Factors', fontsize = 15)
sns.despine()
```

```python
learn_plot = sns.barplot(x="Learning", y="RMSE", data=learning, color = 'blue', saturation=.3)
learn_plot.axes.set_title('RMSE vs. Change in the Learning Rate', fontsize = 15)
learn_plot.set_xlabel("Learning Rate")
sns.despine()
```

```python
reg_plot = sns.barplot(x="Reg", y="RMSE", data=reg, color = 'blue', saturation=.3)
reg_plot.axes.set_title('RMSE vs. Change in the Regularization Rate', fontsize = 15)
reg_plot.set_xlabel("Regularization Rate")
sns.despine()
```
