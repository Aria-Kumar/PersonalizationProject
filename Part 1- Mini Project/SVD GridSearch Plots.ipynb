```python
import os 
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.set(color_codes=True)
```

```python
np.random.seed(sum(map(ord, "distributions")))
```

```python
factors = pd.read_table('factors.csv', header = 0, sep = ',')
learning = pd.read_table('learning.csv', header = 0, sep = ',')
reg = pd.read_table('reg.csv', header = 0, sep = ',')
data_size = pd.read_table('datasize.csv', header = 0, index_col = 1)
```

```python
data_size_RMSE = data_size[data_size['Metric'] == 'RMSE']
data_size_MAE = data_size[data_size['Metric'] == 'MAE']

data_size_RMSE = data_size_RMSE[['Baseline','SVD']]
data_size_MAE = data_size_MAE[['Baseline','SVD']]
```

```python
factors_plot = sns.pointplot(x="Factors", y="RMSE", data=factors)
factors_plot.axes.set_title('RMSE with Change in Factors', fontsize = 15)
sns.despine()

# Saving plots to a local directory
factors_plot = factors_plot.get_figure()
factors_plot.savefig("factors_plot.png")
```

```python
learn_plot = sns.pointplot(x="Learning", y="RMSE", data=learning)
learn_plot.axes.set_title('RMSE with Change in the Learning Rate', fontsize = 15)
learn_plot.set_xlabel("Learning Rate")
sns.despine()

# Saving plots to a local directory
learn_plot = learn_plot.get_figure()
learn_plot.savefig("learn_plot.png")
```

```python
reg_plot = sns.pointplot(x="Regularization", y="RMSE", data=reg)
reg_plot.axes.set_title('RMSE with Change in the Regularization Rate', fontsize = 15)
reg_plot.set_xlabel("Regularization Rate")
sns.despine()

# Saving plots to a local directory
reg_plot = reg_plot.get_figure()
reg_plot.savefig("reg_plot.png")
```

```python
plt.figure(); data_size_RMSE.plot();
plt.xlabel('Number of Artists')
plt.ylabel('RMSE')
plt.savefig('data_size_RMSE.png')
```

```python
plt.figure(); data_size_MAE.plot();
plt.xlabel('Number of Artists')
plt.ylabel('MAE')
plt.savefig('data_size_MAE.png')
```
